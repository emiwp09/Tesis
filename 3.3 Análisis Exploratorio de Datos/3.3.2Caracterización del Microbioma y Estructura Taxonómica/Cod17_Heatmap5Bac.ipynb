```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


top5_bacteria_file_path = r'C:\Users\Emiliano\Desktop\Caracterizacin del Microbioma y Estructura Taxonmica\TOP5bacteriesraces.csv'

top5_bacteria_df = pd.read_csv(top5_bacteria_file_path)

for column in top5_bacteria_df.columns[1:]:
    top5_bacteria_df[column] = pd.to_numeric(top5_bacteria_df[column].replace('%', '', regex=True), errors='coerce')


top5_bacteria_df.set_index('Unnamed: 0', inplace=True)

top5_bacteria_df.loc['TOTAL'] = top5_bacteria_df.sum()

# Crear el mapa de calor con los nombres de las bacterias
plt.figure(figsize=(10, 8))
sns.heatmap(top5_bacteria_df, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Porcentaje (%)'})

# Etiquetas y ttulo
plt.title("Mapa de Calor: TOP 5 Bacterias ms Comunes por Categora Racial")
plt.ylabel("Bacterias")
plt.xlabel("Categoras Raciales")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Mostrar el grfico
plt.show()
```
