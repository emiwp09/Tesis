```python
import pandas as pd

metadata_df = pd.read_csv(r"C:\Users\Emiliano\Desktop\CodigosTesis\Training\Training\metadata\metadata.csv")
taxonomy_df = pd.read_csv(r"C:\Users\Emiliano\Desktop\CodigosTesis\Training\Training\taxonomy\taxonomy_relabd.species.csv")

nih_racial_categories_unique = metadata_df['NIH Racial Category'].unique()

nih_racial_bacteria_summary_filled = {}


for category in nih_racial_categories_unique:
    category_df = metadata_df[metadata_df['NIH Racial Category'] == category]

    filtered_taxonomy_df = taxonomy_df[
        taxonomy_df['specimen'].isin(category_df['specimen'])
        ]
    bacteria_percentages = filtered_taxonomy_df.iloc[:, 1:].mean().fillna(0)

    sorted_bacteria_percentages_top5 = (
        bacteria_percentages.sort_values(ascending=False).head(5)
        )
    
    nih_racial_bacteria_summary_filled[category] = sorted_bacteria_percentages_top5

nih_racial_bacteria_summary_filled_df = pd.DataFrame(nih_racial_bacteria_summary_filled).fillna(0)


output_csv_path = 'TOP5bacteriesraces.csv'
nih_racial_bacteria_summary_filled_df.to_csv(output_csv_path, index=True)

print("Archivo guardado en:", output_csv_path)
```
