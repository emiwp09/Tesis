```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

top10_bacteria_file_path = r'C:\Users\Emiliano\Desktop\Caracterizacin del Microbioma y Estructura Taxonmica\TOP10bacteriesraces.csv'

top10_bacteria_df = pd.read_csv(top10_bacteria_file_path)

for column in top10_bacteria_df.columns[1:]:
    top10_bacteria_df[column] = pd.to_numeric(top10_bacteria_df[column].replace('%', '', regex=True), errors='coerce')#replace('%', '', regex=True): Elimina el smbolo de porcentaje % de los valores en cada columna, usando regex=True para que aplique a todos los valores en cada columna.

top10_bacteria_df.set_index('Unnamed: 0', inplace=True)
top10_bacteria_df.loc['TOTAL'] = top10_bacteria_df.sum()

plt.figure(figsize=(10, 8))
sns.heatmap(top10_bacteria_df, annot=True, cmap="YlGnBu", cbar_kws={'label': 'percentage (%)'})


plt.title("Heatmap: TOP 10 Bacterias ")
plt.ylabel("Bacterias")
plt.xlabel("racial category")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)


plt.show()
```
