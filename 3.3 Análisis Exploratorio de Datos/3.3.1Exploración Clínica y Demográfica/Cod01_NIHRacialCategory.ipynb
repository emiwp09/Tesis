```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\Emiliano\Desktop\CodigosTesis\Training\Training\metadata\metadata.csv'
data = pd.read_csv(file_path)


race_counts = data['NIH Racial Category'].value_counts()

##Construrir la grafica
plt.figure(figsize=(12, 8))
bars = race_counts.plot(kind='bar')
plt.title('NIH Racial Category Chart')
plt.xlabel('Racial Category')
plt.ylabel('Cuantity')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()

for bar in bars.patches:
    bars.annotate(format(bar.get_height(), '.0f'),
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='center', size=10, xytext=(0, 8),
                  textcoords='offset points')


plt.show()
```
