```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = r'C:\Users\Emiliano\Desktop\CodigosTesis\Training\Training\metadata\metadata.csv'
data = pd.read_csv(file_path)

age_race_counts = data.groupby(['age', 'race']).size().unstack(fill_value=0)


age_race_counts.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Age distribution of races ')
plt.xlabel('Ages')
plt.ylabel('quantity')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Race', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()


plt.show()
```
