```python
import pandas as pd

file_path = r'C:\Users\Emiliano\Desktop\CodigosTesis\Training\Training\metadata\metadata.csv'
data = pd.read_csv(file_path)


data['age'] = pd.to_numeric(data['age'], errors='coerce')


participant_age_summary = data.groupby(['NIH Racial Category', 'participant_id']).agg(
    total_known_ages=('age', lambda x: x.notnull().sum()),    
    total_unknown_ages=('age', lambda x: x.isnull().sum())    
).reset_index()

age_summary = participant_age_summary.groupby('NIH Racial Category').agg(
    total_participants=('participant_id', 'nunique'),         
    known_ages=('total_known_ages', lambda x: (x > 0).sum()),   
    unknown_ages=('total_unknown_ages', lambda x: (x > 0).sum()) 
).reset_index()

print(age_summary)

age_summary.to_csv('AgesNIHRacialCategory.csv', index=False)
```
