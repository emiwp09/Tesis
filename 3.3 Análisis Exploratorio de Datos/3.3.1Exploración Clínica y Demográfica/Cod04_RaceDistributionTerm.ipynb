```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\Emiliano\Desktop\CodigosTesis\Training\Training\metadata\metadata.csv'
data = pd.read_csv(file_path)


term_race_counts = data.groupby(['was_term', 'race']).size().unstack(fill_value=0) 


term_race_counts.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Race distribution by Term or was term')
plt.xlabel('was_term')
plt.ylabel('Quantity')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Race', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout() 

plt.show()
```
